# Proyecto 1: Validación de transacciones en E-Commerce
#### Perfil del proyecto: Data Engineer
## Cliente: Mónica Sanelli
## Cargo: Administración y Finanzas y Optimización de procesos
## Empresa: CRT Ltda.
### Responsabilidades:
    - Implementación de plataformas tecnológicas
    - Capacitación de las plataformas en las áreas Administración, 
    - Entrega de información contable
    - Financiera: consolidación bancaria, revisar pagos de Transbank, MercadoPago, PagoProveedores
### Objetivo:
    - Que se les paguen a los trabajadores por el servicio.
    - Mostrar información verídica a SII.

### Proyecto
### Problemática: 
Al realizar una compra, ya sea, crédito o débito por Transbank, no aparece mayor detalle del abono que se realizó (compra), por lo que el usuario debe estar ingresando constantemente a Transbank para obtener los detalles de los abonos, cuotas o devolución.
Actualmente, se debe descargar un informe con todas las transacciones realizadas, y comenzar a ordenar la información manualmente, para luego poder realizar las estadísticas que necesita el usuario.

### Objetivos:
- Automatizar el análisis de la información entregada en el informe de Transbank.
- Generar un "reporte" para el usuario con las salidas que necesita visualizar.

### Beneficios:

- Mayor disponibilidad de tiempo para el usuario para poder realizar otras actividades.


# Proyecto 2: Extraer información para licitaciones de equipos médicos
#### Perfil del proyecto: Public Data Engineer
## Cliente: Rodrigo Pereira
## Cargo: Encargado de Chile
## Empresa: Iberica&Mantenimiento
### Área: Venta de equipos y accesorios médicos
### Responsabilidades:
    - Vender equipos médicos a los hospitales
    - Vender accesorios médicos a los hospitales

### Proyecto
### Problemática: Actualmente se buscan las licitaciones en mercado público de forma manual, revisando el los correos que genera el portal diariamente. Se necesita automatizar este proceso para minimizar los tiempos de entrega de los productos.
### Objetivos:
    - Generar una herramienta que extraiga la información que requiere el cliente automáticamente.
    - Realizar un pequeño análisis sobre la licitación como:
        - Ver cuánto tiempo ha pasado desde la última licitación, con el objetivo de investigar la recurrencia de compra de los hospitales.

### Beneficios:
    - Minimiza los tiempos de entrega de los productos.
    - Se mejora el contacto con los proveedores y los tiempos de importación.
    - Aumenta las posibilidades de ganar una licitación entre un 20% y un 30%.
